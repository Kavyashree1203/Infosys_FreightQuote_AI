## Exact a Table from WebPage

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Set the URL of the website with the table
url = 'https://www.w3schools.com/html/html_tables.asp'

# Make a request to the website and get the HTML response
response = requests.get(url)
html = response.text

# Create a BeautifulSoup object from the HTML response
soup = BeautifulSoup(html, 'html.parser')

# Find the table element containing the data
table = soup.find('table', {'id': 'customers'})

# Check if the table is found before proceeding
if table is not None:
    # Create a list of lists to store the table data
    data = []

    # Extract header row separately
    header_row = table.find('tr')
    header_data = [cell.text.strip() for cell in header_row.find_all(['th', 'td'])]
    data.append(header_data)

    # Iterate over the remaining rows in the table
    for row in table.find_all('tr')[1:]:
        # Create a list to store the data in each row
        row_data = [cell.text.strip() for cell in row.find_all(['th', 'td'])]

        # Add the row data to the list of lists
        data.append(row_data)

    # Create a pandas DataFrame from the list of lists
    df = pd.DataFrame(data[1:], columns=data[0])

    # Print the DataFrame
    print(df.to_string(index=False))
else:
    print("Table not found. Please check the URL or the website's structure.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Set the URL of the website with the table
url = 'https://www.worldometers.info/world-population/population-by-country/'

# Make a request to the website and get the HTML response
response = requests.get(url)
html = response.text

# Create a BeautifulSoup object from the HTML response
soup = BeautifulSoup(html, 'html.parser')

# Find the table element containing the data
table = soup.find('table', {'id': 'example2'})

# Check if the table is found before proceeding
if table is not None:
    # Create a list of lists to store the table data
    data = []

    # Extract header row separately
    header_row = table.find('thead').find('tr')
    header_data = [cell.text.strip() for cell in header_row.find_all('th')]
    data.append(header_data)

    # Iterate over the remaining rows in the table
    for row in table.find('tbody').find_all('tr'):
        # Create a list to store the data in each row
        row_data = [cell.text.strip() for cell in row.find_all('td')]

        # Add the row data to the list of lists
        data.append(row_data)

    # Create a pandas DataFrame from the list of lists
    df = pd.DataFrame(data[1:], columns=data[0])

    # Print the DataFrame
    print(df.to_string(index=False))
else:
    print("Table not found. Please check the URL or the website's structure.")
```
